```python
import os, sys
from pathlib import Path

from md import md

dirDict = dict( 
    llcDIR = os.path.join(Path.home(), 'GDrive/Family/Assets-Hobby/RealEstateInvestments/LLC-WB-Group/pages/'),
    medhcDIR = os.path.join(Path.home(),'GDrive/Family/Health/Medical/HearingCare')
)

docDict = dict(RentalRules = 'llcDIR/IRSGuide/AssetBasedLoan_AlexandraRojas/LLC-IRSGuide-RentalRules.md',
               GiftToAlex = 'llcDIR/IRSGuide/AssetBasedLoan_AlexandraRojas/GiftToAlex.md',
               hearing = 'medhcDIR/20251222_Hearing.md'
              )
p = docDict['hearing']
k = p.split('/')[0]
DIR = dirDict[k]


FN = os.path.join(DIR, p.replace(f"{k}",DIR))
print("Using FN:", FN)
with open(FN,'r') as fio:
    mdStr = fio.read()

m = md(mdStr)
h = m.html()
pdfStr = m.pdf(html = h)

outFN = FN.replace('.md','.pdf')
print("Saving outFN:", outFN)
with open(outFN,'wb') as fio:
    fio.write(pdfStr)
```
