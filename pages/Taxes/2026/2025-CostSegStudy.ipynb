- **Ownership Matters**: You must take personal ownership of the property (not hold it within a Self-Directed IRA) to claim the depreciation deduction.

### Turbo Tax K1 : Move LLC deducations -> Personal Deducations 

K-1 losses may not flow into your personal deductions due to specific IRS rules:

1. **Adjusted Basis in LLC**: You can only deduct losses up to your adjusted basis in the partnership or S corporation.

    - based on your K-1 depends primarily on your adjusted basis in the LLC.
    - **Adjusted basis in LLC**
        - original investment,
        - increased by your share of income
        - decreased by your share of losses and distributions.
        - K-1 :  member’s beginning basis, income, losses, and distributions 
    - can deduct losses and claims such as Section 179 up to the amount of your adjusted basis.
    - if deducations > adjusted basis :: you can not move deducations into your personal return, for current yearStill
    - Still, annual deductions can often be carried forward to future years when your basis increases.

3. **At-Risk Limitations**:

    - Deductions are limited to the amount you have "at risk" in the business,
    - meaning your actual economic investment.

5. **Passive Activity Loss Rules**:

    - If you do not materially participate, passive losses can only offset passive income, not other income.
    - There is the $25,000 loss limit for non-real estate professionals
        - This limits passive loss passthru if your Adjusted Gross Income > $100K

7. **Suspended Losses**:

    - Losses that exceed these limits are suspended and carried forward to future years.

9. **Non-Deductible Losses**:

    - Certain losses, like those from related-party transactions, are not deductible.

## Key Considerations for Cost Segregation on Long-Term Rentals:
- **Property Eligibility**: The property must be an income-producing investment (rental), not your primary residence. It is best suited for properties purchased or built within the last 15 years.
- **Benefits**: Cost segregation increases upfront depreciation deductions, which can create significant tax losses.
- **Passive Loss Rules**: For long-term rentals, these deductions are generally considered "passive losses". They can offset income from other passive investments but typically cannot offset W-2 income unless you qualify as a Real Estate Professional (REPS).
- **Retroactive Filing**: You can perform a study on a property you have owned for years, usually by filing Form 3115 with your tax return to catch up on missed deductions. 
- For maximum benefit, the cost segregation study should typically be done on properties with a purchase price/basis over $300,000.
