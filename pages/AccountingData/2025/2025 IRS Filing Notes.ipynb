## IRS Installment Plans 

Penalty Savings: While the 4-month plan has no setup fee, it carries a higher monthly penalty (0.5%). The 12 and 24-month plans qualify for a reduced penalty (0.25%), but the longer duration leads to higher total interest costs.
Total Savings: Choosing the 12-month plan instead of the 24-month plan saves you approximately $1,960 in total costs. 


#### 1. Short-Term Payment Plan (180 Days)
- **Term**: Up to 180 days (approx. 6 months).
- **Interest Rate**: 6% for the current quarter (April–June 2026), compounded daily.
- **Setup Fee**: $0 (free to set up).
- **Eligibility**: Generally available for individuals owing less than $100,000 in combined tax, penalties, and interest.

#### 2. Long-Term Installment Agreement (Monthly)
- **Term**: Up to 72 months (6 years) standard, though some plans can extend up to the 10-year collection statute.
- **Interest Rate**: 6% annually, compounded daily.
- **Setup Fees (2026)**:
    - $22: Online setup with direct debit (the IRS Direct Debit Installment Agreement).
    - $107: Mail, phone, or in-person setup with direct debit.
    - $69: Online setup without direct debit.
    - $178: Mail, phone, or in-person setup without direct debit.
- **Eligibility**: Individuals owing $50,000 or less can typically apply for a Simple Payment Plan online.

#### 3. Key Penalties & Reductions
- **Late Payment Penalty**: Standard at 0.5% per month.
- **In-Plan Reduction**: If you file on time and enter an installment agreement, this penalty drops to 0.25% per month while the plan is active.
- **Low-Income Options**: Fees may be waived or reimbursed (typically $0 to $43) for taxpayers who meet IRS low-income criteria.
