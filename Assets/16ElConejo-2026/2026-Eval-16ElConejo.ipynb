```python
# Interest Rate history data
csvIntRates = '''Year fyLow - fyHigh desc
2016	3.65% - 3.7%	Historically low
2017	3.98% - 4.1%	Steady
2018	4.5% - 4.6%	Peak of the late 2010s
2019	3.9% - 4.0%	Dip
2020	2.9% - 3.1%	Pandemic low
2021	2.65% - 2.96%	All-time low
2022	5.3% - 5.5%	Mid-year surge
2023	6.5% - 7.5%	High rates:7.79% peak
2024	6.2% - 7.0%	Elevated
2025	6.0% - 6.8%	Slight easing
2026 	5.8% - 6.06% Stabilization'''
def intRateDF(csv):
    newCSV = ''
    for ln in csv.split('\n'):
        wList = ln.replace('-','').replace('%','').split()
        desc = ' '.join(wList[3:])
        newLn = ','.join(wList[0:3] + [desc])
        newCSV += newLn + '\n'
    with StringIO(newCSV) as sio:
        df = pd.read_csv(sio)
    return df
intDF = intRateDF(csvIntRates)
ax = intDF.set_index('Year', drop=True)[['fyLow','fyHigh']].plot(title='Int Rate History', ylabel='Int Rate')
```

## 10 year ROI : RE vs MKT

```python
# Compute ROI
fixup = 10000
salePrice = (askPrice * 0.90) - fixup

mktRate = 0.08
roiYears = 10

def apprValue(PV, R, years):
    p = PV
    for i in range(years):
        p = p * (1.0 + R)

    return int(p)

def fvDF(salePrice, nm, **kwargs):

    rent = kwargs.get('rent', 2600)
    expPct = kwargs.get('expPct', 0.20) # Rental PropTax/Maintainance Expense

    mktFV = apprValue(salePrice, mktRate, roiYears)
    reInc = int(rent*12*roiYears*(1-expPct))
    reFVProp = apprValue(salePrice, reRate, roiYears)
    reFV = reFVProp + reInc
    return pd.DataFrame([dict(SalePrice = int(salePrice), 
                        FV_mkt_8Pct = mktFV, 
                        FV_RE = reFV,
                        FV_mkt_10Pct = apprValue(salePrice, 0.1, roiYears),
                        FV_RE_Prop = reFVProp,
                        FV_RE_Income = reInc,
                        PerYr_mkt = int(mktFV/roiYears), 
                        PerYr_RE = int(reFV/roiYears)
                       )], index=[nm])
roiDF = pd.concat([fvDF(240000,'sale240K'),
           fvDF(265000, 'Sale265K'),
           fvDF(277000, 'Sale277K')
          ]).transpose()
roiDF
```

```python
import matplotlib.pyplot as plt

vDF = roiDF.iloc[:-4]

xList = list(vDF.index)
xLen = len(xList)

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4)) #


vDF.plot(ax=ax1, title='ROI for different sale prices')

ax1.set_xticks(range(xLen))
_ = ax1.set_xticklabels(xList, rotation=45) # Optional: rotate labels for readability

roiDF.iloc[4:6].transpose().plot(ax=ax2, kind='bar', stacked='True', title='Rental: Appreciation/Income')
```
